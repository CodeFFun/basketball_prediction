```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cleaned_shots_logs.csv")
```

```python
df.head()
```

```python
df.info()
```

Before modelling the data we need to understand what features should be used and what result we are expecting. We are trying to predict which shots will be made under certain conditions. For that we have two features`FGM` and `SHOT_RESULT`. `FGM` is already in categorical form of 0/1; however `SHOT_RESULT` isn't. So we will take `FGM` as the result. We will only be working with numeric features, excluding `player_id` and `game_id` since they have no use in prediction. We will also not be taking any categorical data since they don't contribute to the final result. Now that features have been selected we will need to see the distribution of the result.

```python
df["FGM"].value_counts(normalize=True)
```

```python
x_column = ["CLOSE_DEF_DIST", "SHOT_NUMBER", "PERIOD", "GAME_CLOCK", "SHOT_CLOCK", "DRIBBLES", "TOUCH_TIME","SHOT_DIST"]

df_model = df[x_column]
```

The result is distributed in a balanced format, so we will not be conducting any under-- or oversampling of the data. 

### Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
#base logistic model
X = df_model
y = df["FGM"]
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.3, random_state=18)
#train data
log_model = LogisticRegression()
log_model.fit(X_train, y_train)
log_model.predict(X_train)
#test data
y_base_pred = log_model.predict(X_test)
score = log_model.score(X_test,y_test)
print(score)
print(classification_report(y_test,y_base_pred))
```

We have pretty decent results. The total accuracy of the model is 61%. The model predicted 62% of missed shots correctly and 59% of made shots. It also predicted 73% of missed shots accurately, while it only predicted 46% of made shots accurately. Now let's try to fine-tune the model.

```python
from sklearn.preprocessing import StandardScaler

X = df_model
y = df["FGM"]
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.3, random_state=18)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
#train data
log_model_scaled = LogisticRegression(class_weight="balanced", max_iter = 1000)
log_model_scaled.fit(X_train_scaled, y_train)
log_model_scaled.predict(X_train_scaled)
#test data
y_scaled_pred = log_model_scaled.predict(X_test_scaled)
train_score = log_model_scaled.score(X_train_scaled, y_train)
test_score = log_model_scaled.score(X_test_scaled,y_test)
print(train_score, test_score)
print(classification_report(y_test,y_scaled_pred))
```

# Random Forest and Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
X = df_model
y = df["FGM"]
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.3, random_state=18)
#train data
dtree_model = DecisionTreeClassifier(criterion="gini", splitter="best", max_depth=3, random_state=18)
dtree_model.fit(X_train,y_train)
train_pred = dtree_model.predict(X_train)
train_model_prediction = classification_report(y_train, train_pred)
print("Train data Model")
print(train_model_prediction)
#test data
test_pred = dtree_model.predict(X_test)
test_model_prediction = classification_report(y_test, test_pred)
print("Test data model")
print(test_model_prediction)
```

The result are similar to the logistic regression base model. We also predicted both train data and test data in order to see if overfitting exist but it seems there is no oiverfitting. Now we will see the graph of how decision was made and then optimize the model

```python
from sklearn.tree import plot_tree

plt.figure(figsize = [24.0, 8.0])

_ = plot_tree(dtree_model, 
          feature_names = X.columns, 
          class_names=[str(c) for c in dtree_model.classes_],
          filled = True,        
          proportion = False,
          precision = 3,
          rounded = True,
          fontsize = 11) 

plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
X = df_model
y = df["FGM"]
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.3, random_state=18)
tree_scaler = StandardScaler()
X_train_scaled = tree_scaler.fit_transform(X_train)
X_test_scaled = tree_scaler.transform(X_test)
#train data
param_grid = {
    "max_depth": [3, 5, 8, 10, 15, None],
    "min_samples_split": [2, 5, 10, 20],
    "min_samples_leaf": [1, 2, 5, 10],
    "criterion": ["gini", "entropy"],
    "max_features": [None, "sqrt", "log2"]
}

grid_search = GridSearchCV(
    DecisionTreeClassifier(class_weight="balanced", random_state=18),
    param_grid,
    cv=5,                    # 5-fold cross validation
    scoring="f1_weighted",   # good for imbalanced classes
    n_jobs=-1,
    verbose=1
)

grid_search.fit(X_train_scaled, y_train)
print("Best Params:", grid_search.best_params_)
print("Best CV Score:", grid_search.best_score_)
best_model = grid_search.best_estimator_

# dtree_model_scaled = DecisionTreeClassifier(criterion="gini", splitter="best", max_depth=8, random_state=18, class_weight="balanced")
# dtree_model_scaled.fit(X_train_scaled, y_train)
# dtree_model_scaled.predict(X_train_scaled)
# #test data
# y_scaled_pred = dtree_model_scaled.predict(X_test_scaled)
# train_score = dtree_model_scaled.score(X_train_scaled, y_train)
# test_score = dtree_model_scaled.score(X_test_scaled,y_test)
# print(train_score, test_score)
# print(classification_report(y_test,y_scaled_pred))
```
