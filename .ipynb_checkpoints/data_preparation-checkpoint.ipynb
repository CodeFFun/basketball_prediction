### Data Preparation
Now lets clean some data, since `SHOT_CLOCK` had some null values. Making the assumptions that the `SHOT_CLOCK` is equal to `GAME_CLOCK`, we will replace the NA value with the value of `GAME_CLOCK` when `GAME_CLOCK` is less than 25 seconds, since a NBA offense is of 24 seconds. And we will also be changing the game clock to `seconds` format instead of `minute:second`.

```python
cleanDF = df.copy() 
```

```python
cleanDF["GAME_CLOCK"] = cleanDF["GAME_CLOCK"].apply(lambda x: int(x.split(":")[0]) * 60 + int(x.split(":")[1]))
null_mask = cleanDF["SHOT_CLOCK"].isnull()
cleanDF.loc[(cleanDF["SHOT_CLOCK"].isnull()) & (cleanDF["GAME_CLOCK"] < 25), "SHOT_CLOCK"] = cleanDF["GAME_CLOCK"]
cleanDF = cleanDF[~cleanDF['SHOT_CLOCK'].isnull()]
```

```python
cleanDF["SHOT_CLOCK"].head(30)
```

```python
cleanDF.loc[null_mask, ["SHOT_CLOCK", "GAME_CLOCK"]]
```

```python
cleanDF[["player_name", "CLOSEST_DEFENDER"]]
```

Here there is no consistencty in players name. Shooters have  small case and is formatted like `firstname` `lastname`, while defenders are formatted as `lastname`, ',',`firstname`. with first letter being capital. So we will format them by making them all capital and name format will be `lastname`, `comma`, `firstname`.

```python
def formatName(data):
    firstname = data.split(' ')[0]
    lastname = data.split(' ')[1]
    fullname = f'{lastname}, {firstname}'
    return fullname
```

```python
shooter_name = cleanDF["player_name"]
shooter_name = shooter_name.apply(lambda x : x.upper())
cleanDF["player_name"] = shooter_name.apply(formatName)
cleanDF["player_name"].head()
```

```python
cleanDF["CLOSEST_DEFENDER"] = cleanDF["CLOSEST_DEFENDER"].str.upper()
cleanDF["CLOSEST_DEFENDER"] = cleanDF["CLOSEST_DEFENDER"].str.replace(".", "", regex=False)
cleanDF["CLOSEST_DEFENDER"].head()
```

```python
cleanDF.loc[cleanDF['TOUCH_TIME'] <= 0.1, "TOUCH_TIME"]
```

Realistically if a player is s|hooting after the shot clock is less than 0 than that means the data is incorrect or doesn't count. Those data are either wrong or are of no need so all of these data are thrown.

```python
cleanDF = cleanDF[cleanDF['TOUCH_TIME'] >= 0.1]
cleanDF['TOUCH_TIME'].min()
```

Now that we have cleaned the data properly lets list out what we have changes so we can keep on generating clean data
- First we converted `GAME_CLOCK` into seconds from M:S fromat.
- Then the `SHOT_CLOCK` with `NaN` values were replaced with `GAME_CLOCK` when `GAME_CLOCK` was less than 25 seconds.
- Then the names were formatted in `LASTNAME, FIRSTNAME` fromat and the dots were removed from the name.
- Finnaly we removed all the rows where `TOUCH_TIME` were less than 0.1

For further analysis and to see how the features correlate with each other, we will use unsupervised learning to create the cluster and understand the relations.

```python
from sklearn.cluster import KMeans
```

```python
numeric_columns = ["SHOT_NUMBER", "PERIOD", "GAME_CLOCK", "SHOT_CLOCK", "DRIBBLES", "TOUCH_TIME", "SHOT_DIST", "CLOSE_DEF_DIST"]
```

```python
for i in numeric_columns:
    print(f'{i} : {cleanDF[i].isnull().sum()}')
```

```python
def plot_elbow_curve(df, max_clusters=15):
    inertias = []
    for k in range(1, max_clusters+1):
        model = KMeans(n_clusters = k, random_state=744)
        cluster = model.fit_predict(df)
        inertias.append(round(model.inertia_,2))
    plt.figure(figsize=(16,6))
    plt.plot(range(1, max_clusters+1), inertias, marker = 'x')
    plt.xticks(ticks = range(1,max_clusters+1), labels=range(1,max_clusters+1))
    plt.title("Inertia vs Number of Clusters")

    plt.tight_layout()
    plt.show()
    return inertias
```

```python
inertias = plot_elbow_curve(cleanDF[numeric_columns])
print(inertias)
```

From the above elbow graph, we can see that the drop-off is low when the cluster goes from 3 to 4. But just to be sure we will use BaysianGaussianMixture to confirm the clusters from `1 to 6 since from 6 there is clearly less to no value change.

```python
from sklearn.mixture import BayesianGaussianMixture
```

```python
for i in range(1,6):
    model = BayesianGaussianMixture(n_components = i, random_state = 744)
    model.fit_predict(cleanDF[numeric_columns])
    print(f"n={i} | weights: {model.weights_.round(3)}")
```

On default settings we can see that every cluster has only 3 high-weight values, which can confirm that 3 is the definitive cluster. However, the model was run only on base values and after 3 it is showing error, we can try to tweak the tol, max_iter, or init state however above model took quite some time and a lot of resource and since both k-means and  Bayessian has given 3 cluster as common point we will take 3 cluster as the answer.

```python
def cluster_diagram(df, clusters):
    kmeans = KMeans(n_clusters = clusters, random_state-41)
    cluster = kmeans.fit_predict(df)
    df_copy = df.copy()
    df_copy["cluster"] = cluster
    plt.subplot(figsize=(12,8))
    plt.plot()
    plt.show()
```
